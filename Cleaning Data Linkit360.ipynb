```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np


columns = ['revenue', 'service_price', 'attempt_charging', 'success_billing']

plt.figure(figsize=(10, 10))
for i, col in enumerate(columns):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot for {col}', fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
for i, col in enumerate(columns):
    plt.subplot(2, 2, i + 1)
    plt.hist(data[col], bins=20, rwidth=0.8, density=True)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)

    rng = np.arange(data[col].min(), data[col].max(), 0.1)
    plt.plot(rng, norm.pdf(rng, data[col].mean(), data[col].std()))
    plt.title(f'Histogram and Normal Distribution for {col}', fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

### **Handling Outlier**

deteksi outlier

```python
# columns = ['revenue', 'service_price', 'attempt_charging', 'success_billing']

for col in columns:
    percentile25 = data[col].quantile(0.25)
    percentile75 = data[col].quantile(0.75)
    iqr = percentile75 - percentile25

    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr

    print(f"Lower limit for {col}:", lower_limit)
    print(f"Upper limit for {col}:", upper_limit)

    outliers = data[(data[col] > upper_limit) | (data[col] < lower_limit)]

    print(f"Number of rows in the data containing outliers in {col}:", len(outliers))
    print()
```

nah disini kalo data outlier kita hapus nah kalo milih di hapus maka pake "data no outier" ini saja, kalo mau di pertahankan maka pake variable "data"

```python
data_no_outliers = data[(data[col] <= upper_limit) & (data[col] >= lower_limit)]
data_no_outliers.to_csv('data_no_outliers.csv', index=False)
```
